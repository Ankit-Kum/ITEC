**Calculate trainable parameters for sales value predictions**

```python
import pandas as pd
df = pd.read_csv('/content/Adv.csv',index_col=0)
df.head()
```

```python
df.shape
```

```python
x = df['TV']
y = df['Sales']
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.show()
```

```python
w0 = 2
w1 = 2.1
yh = w0 + w1 * x

error = ((y - yh)**2).mean()
print(error)
```

```python
dew0 =  -2 * ((y - yh)).mean()
dew1 =  -2 * ((y - yh) * x).mean()

print(dew0 , dew1)
```

```python
lr = 1e-5

w0 = w0 - lr * dew0
w1 = w1 - lr * dew1

yh = w0 + w1 * x

error = ((y - yh)**2).mean()
print(error)
```

```python
w0 = 7
w1 = 6
for i in range(0,50):
  yh = w0 + w1 * x

  dew0 =  -2 * ((y - yh)).mean()
  dew1 =  -2 * ((y - yh) * x).mean()

  lr = 0.00001

  w0 = w0 - lr * dew0
  w1 = w1 - lr * dew1

  error = ((y - yh)**2).mean()
  print(i,error)
```

```python
print(w0, w1)
```

```python
mae = abs(y - yh).mean()
print(mae)
```

```python
plt.scatter(x,y)
plt.plot(x,yh,'r')

plt.show()
```

```python
tv_new = 56.2
sales_predicted = w0 + w1 * tv_new
print(sales_predicted)
```

**Multiple features or Multi linear Regression**

```python
x = df[['TV','Radio']]
y = df['Sales']
```

```python
#yh = w0 + w1*x1 + w2*x2
#yh = w0 + w1*tv + w2*radio

w0 = 4
w1 = 3
w2 = 2

yh = w0 + w1 * x.TV + w2 * x.Radio
error = ((y - yh)**2).mean()
print(error)
```

```python
#dew0 = -2*((y-yh)).mean()
#dew1 = -2*((y-yh)*(x.tv)).mean()
#dew2 = -2*((y-yh)*(x.radio)).mean()
```

```python
w0 = 4
w1 = 3
w2 = 2

for i in range(0,1000):
  yh = w0 + w1 * x.TV + w2 * x.Radio

  dew0 = -2*((y-yh)).mean()
  dew1 = -2*((y-yh)*(x.TV)).mean()
  dew2 = -2*((y-yh)*(x.Radio)).mean()

  lr = 0.00001

  w0 = w0 - lr * dew0
  w1 = w1 - lr * dew1
  w2 = w2 - lr * dew2

  error = ((y - yh)**2).mean()
  print(i,error)
```

```python
mae = abs(y - yh).mean()
print(mae)
```

**Using Radio as a feature to train my sales prediction model improves the performance by reducing the error value**

```python
tv_new = 56.2
radio_new = 67.8

sales_predicted = w0 + w1 * tv_new + w2 * radio_new
print(sales_predicted)
```
