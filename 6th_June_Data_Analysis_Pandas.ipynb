```python
### Handling Null Values
```

```python
## Count of null values
```

```python
data.isnull().sum()
```

```python
### Drop rows or columns having null values
```

```python
data_1 = data.dropna(axis=0)  ## drop all rows which are having null values
data_1.shape
```

```python
data_1 = data.dropna(axis=1)  ## drop all columns which are having null values
data_1.shape
```

```python
data_1 = data.dropna(axis=0,how='any')  ## drop all rows where any column is having null values
data_1.shape
```

```python
data_1 = data.dropna(axis=0,how='all')  ## drop all rows where all column is having null values
data_1.shape
```

```python
data_1 = data.dropna(axis=0,subset=['Cabin'])  ## drop all rows where all Cabin Column is having null values
data_1.shape
```

```python
data_1 = data.dropna(axis=0,subset=['Embarked'])  ## drop all rows where all Embarked Column is having null values
data_1.shape
```

```python
### fill null values
```

```python
mean_age = data['Age'].mean()
mean_age = round(mean_age,2)
mean_age
data['Age'] = data['Age'].fillna(mean_age)
```

```python
data['Cabin'] = data['Cabin'].fillna("NA")
```

```python
data['Embarked'].value_counts()
```

```python
data['Embarked'] = data['Embarked'].fillna("S")
```
