**Logistic Regression** - An algorithm used for categorical value prediction

```python
import numpy as np
data = np.array([[3.87,2.65,0],
                 [2.64,3.63,0],
                 [4.93,5.50,0],
                 [2.83,2.95,0],
                 [4.07,4.00,0],
                 [8.27,3.57,1],
                 [6.36,3.09,1],
                 [7.99,2.97,1],
                 [9.76,1.42,1],
                 [8.76,4.60,1]])
print(data)
```

```python
from math import exp

def predict_val(data_row,coef):
  yh = coef[0] + coef[1] * data_row[0] + coef[2] * data_row[1]
  return 1 / (1 + exp(-yh))
```

```python
coef = [0.9,0.8,1.1]

for row in data:
  print(row[2] , predict_val(row,coef), round(predict_val(row,coef)))
```

```python
#Stocastic gradient descent

for iter in range(500):
  sum_error = 0
  for row in data:
    #error = y - z
    z = predict_val(row,coef)
    error = row[2] - z
    sum_error += error**2

    lr = 1

    #grad = -2 * (y-z) * z * (1-z) * x
    #w = w - lr * de/dw

    coef[0] = coef[0] - lr * (-2 * error * z * (1-z))
    coef[1] = coef[1] - lr * (-2 * error * z * (1-z) * row[0])
    coef[2] = coef[2] - lr * (-2 * error * z * (1-z) * row[1])

  print(iter,sum_error)
```

```python
print(coef)
```

```python
for row in data:
  print(row[2] , ' ***  ' , predict_val(row,coef), ' *** ' , round(predict_val(row,coef)))
```

```python
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(random_state=1)
X_res, y_res = ros.fit_resample(x, y)
```

```python
#Evaluate

accTrain = (ytrain == model1.predict(xtrain)).sum()/len(xtrain)
print(accTrain)
```

```python
model1.score(xtrain,ytrain)
```

```python
accTest = model1.score(xtest,ytest)
print(accTest)
```

**Customer Churn Classifications**

```python
import numpy as np
import pandas as pd
df = pd.read_csv('/content/customer_churn.csv')
df.head()
```

```python
df1 = df.drop('customerID' , axis = 'columns')
df1.head()
```

```python
df1.isnull().sum()
```

```python
df1.dtypes
```

```python
def fix_totalcharges(x):
  try:
    return float(x)
  except:
    return np.nan
```

```python
df1['TotalCharges'] = df1['TotalCharges'].apply(fix_totalcharges)
```

```python
df1.isnull().sum()
```

```python
df1.shape
```

```python
df2 = df1.dropna(axis='rows')
df2.head()
```

```python
df2.dtypes
```

```python
non_num = []
for col in df2.columns:
  if (df2[col].dtype == object):
    non_num.append(col)
```

```python
print(non_num)
```

```python
for cols in non_num:
  print(cols , df2[cols].unique())
```

```python
def fix_yes_no(x):
  if x == 'Yes':
    return 'Yes'
  else:
    return 'No'
```

```python
df2['MultipleLines'] = df2['MultipleLines'].apply(fix_yes_no)
df2['OnlineSecurity'] = df2['OnlineSecurity'].apply(fix_yes_no)
df2['OnlineBackup'] = df2['OnlineBackup'].apply(fix_yes_no)
df2['DeviceProtection'] = df2['DeviceProtection'].apply(fix_yes_no)
df2['TechSupport'] = df2['TechSupport'].apply(fix_yes_no)
df2['StreamingTV'] = df2['StreamingTV'].apply(fix_yes_no)
df2['StreamingMovies'] = df2['StreamingMovies'].apply(fix_yes_no)
```

```python
for cols in non_num:
  print(cols , df2[cols].unique())
```

```python
for cols in non_num:
  print(cols , df2[cols].value_counts())
  print('******************')
```

```python
df2.head()
```

```python
#One hot encoding
pd.get_dummies(df2['gender'])
```

```python
non_num_data = df2[non_num].drop('Churn' , axis='columns')
non_num_data.head()
```

```python
encoded_non_num = pd.get_dummies(non_num_data)
encoded_non_num.head()
```

```python
df2.head()
```

```python
df3 = df2.drop(non_num , axis='columns')
df3.head()
```

```python
X = pd.concat((df3,encoded_non_num) , axis = 'columns')
X.head()         #Features
```

```python
Y = df2['Churn'].map({'No' :0 , 'Yes' : 1})
Y
```

```python
Y.value_counts()
```

```python
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(random_state=1)
X_res, Y_res = ros.fit_resample(X, Y)
```

**Split, Import algorithm and train the model**

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(X_res , Y_res , train_size=0.80)

model2 = LogisticRegression()

model2.fit(xtrain,ytrain)
```

```python
print(model2.score(xtrain,ytrain))
print(model2.score(xtest,ytest))
```

```python
confusion_matrix(ytrain , model2.predict(xtrain))
```

```python
confusion_matrix(ytest , model2.predict(xtest))
```
