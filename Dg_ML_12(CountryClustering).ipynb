```python
import pandas as pd
df = pd.read_csv('/content/Country-data.xls')
df.head()
```

```python
df1 = df.drop('country' , axis=1)
df1.head()
```

```python
from sklearn.cluster import KMeans
```

```python
losses = []
for i in range(1,12):
  model = KMeans(n_clusters=i)
  model.fit(df1)
  losses.append(model.inertia_)

print(losses)
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1,12), losses)
plt.show()
```

```python
df2 = df1.copy()
```

```python
model1 = KMeans(n_clusters = 3)
model1.fit(df2)

pd.DataFrame(model1.cluster_centers_ , columns = df2.columns)
```

```python
df['labels'] = model1.predict(df2)
```

```python
df.head()
```

```python
cluster0 = df[df['labels'] == 0]
```

```python
cluster0.head()
```

```python
cluster0['country']
```
