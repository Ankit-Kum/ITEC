```python
import pandas as pd
df = pd.read_csv('/content/Call Center Data.csv')
df.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['Experience in months'],df['Call Attended in a day'])
plt.show()
```

```python
from sklearn.cluster import KMeans
model1 = KMeans(n_clusters = 4)
model1.fit(df)
```

```python
df['Label'] = model1.predict(df)
```

```python
df.head()
```

```python
pd.DataFrame(model1.cluster_centers_ , columns = ['Experience in months','Call Attended in a day'])
```

```python
#Cluster2 members
df[df.Label == 2]
```

```python
plt.scatter(df['Experience in months'],df['Call Attended in a day'],
            c = df['Label'])
plt.show()
```
