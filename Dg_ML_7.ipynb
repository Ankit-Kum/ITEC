**Decision Tree Classifier**

```python
from sklearn.datasets import load_wine
wine = load_wine()

wine
```

```python
wine.keys()
```

```python
wine['feature_names']
```

```python
import pandas as pd
import numpy as np

X = pd.DataFrame(wine['data'] , columns = wine['feature_names'])
X.head()
```

```python
wine['target_names']
```

```python
Y = wine['target']
Y
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(X,Y,random_state=1)
```

```python
from sklearn.tree import DecisionTreeClassifier
model1 = DecisionTreeClassifier()
```

```python
model1.fit(xtrain,ytrain)
```

```python
print(model1.score(xtrain,ytrain))
print(model1.score(xtest,ytest))
```

* Differnce between train and test accuracy is huge so we call it a overfit model which means model can perform well on known information but it will be lesser accurate for unknown sets of information
* We need to tune model so that overfitting can be removed

```python
from sklearn.tree import export_graphviz
from IPython.display import Image
import pydotplus

dot_data = export_graphviz(model1,feature_names=xtrain.columns)
graph = pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
```

**Limiting Criteria of tree formed**

* max_depth

```python
from sklearn.tree import DecisionTreeClassifier
model2 = DecisionTreeClassifier(max_depth = 2)
model2.fit(xtrain,ytrain)

print(model2.score(xtrain,ytrain))
print(model2.score(xtest,ytest))
```

```python
dot_data = export_graphviz(model2,feature_names=xtrain.columns)
graph = pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
```

* min_samples_leaf

```python
from sklearn.tree import DecisionTreeClassifier
model3 = DecisionTreeClassifier(min_samples_leaf=11)
model3.fit(xtrain,ytrain)

print(model3.score(xtrain,ytrain))
print(model3.score(xtest,ytest))
```

```python
dot_data = export_graphviz(model3,feature_names=xtrain.columns)
graph = pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
```

* min_samples_split, max_depth

```python
from sklearn.tree import DecisionTreeClassifier
model4 = DecisionTreeClassifier(min_samples_split=10,max_depth=2)
model4.fit(xtrain,ytrain)

print(model4.score(xtrain,ytrain))
print(model4.score(xtest,ytest))
```

```python
dot_data = export_graphviz(model4,feature_names=xtrain.columns)
graph = pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
```

* max_leaf_nodes, min_impurity_decrease

```python
from sklearn.tree import DecisionTreeClassifier
#model5 = DecisionTreeClassifier(max_leaf_nodes=3)
model5 = DecisionTreeClassifier(min_impurity_decrease=0.1,max_leaf_nodes=4,min_samples_leaf=11)
model5.fit(xtrain,ytrain)

print(model5.score(xtrain,ytrain))
print(model5.score(xtest,ytest))
```

```python
dot_data = export_graphviz(model5,feature_names=xtrain.columns)
graph = pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
```
