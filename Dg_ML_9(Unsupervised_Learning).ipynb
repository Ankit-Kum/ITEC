**UnSupervised Learning**
* WHen the data samples are having only featurs.
* We used unlabeled data to label them
* We can use clustering where we identify data samples with similar features and put them into one cluster

```python
import pandas as pd
import numpy as np

df=pd.DataFrame({'x':[12,24,28,33,18,29,52,45,24,55,51,61,53,69,72,64,49,58],
                  'y':[36,39,30,52,54,46,55,59,63,70,66,63,58,23,14,8,19,7]})

df
```

```python
import matplotlib.pyplot as plt
plt.scatter(df.x , df.y)
plt.show()
```

* Define centroids within the minimum and maximum values of given coordinates

```python
print(df['x'].min(),df['x'].max())
print(df['y'].min(),df['y'].max())
```

```python
cent0 = [np.random.randint(12,72),np.random.randint(7,70)]
cent1 = [np.random.randint(12,72),np.random.randint(7,70)]
cent2 = [np.random.randint(12,72),np.random.randint(7,70)]

print(cent0 , cent1 , cent2)
```

```python
plt.scatter(df.x , df.y)

plt.scatter(cent0[0] , cent0[1] , c ='r' , marker='*')
plt.scatter(cent1[0] , cent1[1] , c ='g' , marker='*')
plt.scatter(cent2[0] , cent2[1] , c ='b' , marker='*')

plt.show()
```

```python
df['r'] = np.sqrt((df.x - cent0[0])**2 + (df.y - cent0[1])**2)
df['g'] = np.sqrt((df.x - cent1[0])**2 + (df.y - cent1[1])**2)
df['b'] = np.sqrt((df.x - cent2[0])**2 + (df.y - cent2[1])**2)
```

```python
df.head()
```

```python
df['label'] = df.loc[: , 'r':'b'].idxmin(axis=1)   #closest centeroid for each data points
```

```python
df.head()
```

```python
plt.scatter(df.x , df.y , c = df['label'])

plt.scatter(cent0[0] , cent0[1] , c ='r' , marker='*')
plt.scatter(cent1[0] , cent1[1] , c ='g' , marker='*')
plt.scatter(cent2[0] , cent2[1] , c ='b' , marker='*')

plt.show()
```

```python
cluster_points_avg = df.groupby('label')['x','y'].mean()
cluster_points_avg
```

```python
cent0 = list(cluster_points_avg.loc['r'])
cent1 = list(cluster_points_avg.loc['g'])
cent2 = list(cluster_points_avg.loc['b'])

print(cent0, cent1 , cent2)
```

```python
plt.scatter(df.x , df.y , c = df['label'])

plt.scatter(cent0[0] , cent0[1] , c ='r' , marker='*')
plt.scatter(cent1[0] , cent1[1] , c ='g' , marker='*')
plt.scatter(cent2[0] , cent2[1] , c ='b' , marker='*')

plt.show()
```

```python
df['r'] = np.sqrt((df.x - cent0[0])**2 + (df.y - cent0[1])**2)
df['g'] = np.sqrt((df.x - cent1[0])**2 + (df.y - cent1[1])**2)
df['b'] = np.sqrt((df.x - cent2[0])**2 + (df.y - cent2[1])**2)

df['label'] = df.loc[: , 'r':'b'].idxmin(axis=1)
```

```python
df.head()
```

```python
plt.scatter(df.x , df.y , c = df['label'])

plt.scatter(cent0[0] , cent0[1] , c ='r' , marker='*')
plt.scatter(cent1[0] , cent1[1] , c ='g' , marker='*')
plt.scatter(cent2[0] , cent2[1] , c ='b' , marker='*')

plt.show()
```

```python
cluster_points_avg = df.groupby('label')['x','y'].mean()

cent0 = list(cluster_points_avg.loc['r'])
cent1 = list(cluster_points_avg.loc['g'])
cent2 = list(cluster_points_avg.loc['b'])

print(cent0, cent1 , cent2)
```

```python
plt.scatter(df.x , df.y , c = df['label'])

plt.scatter(cent0[0] , cent0[1] , c ='r' , marker='*')
plt.scatter(cent1[0] , cent1[1] , c ='g' , marker='*')
plt.scatter(cent2[0] , cent2[1] , c ='b' , marker='*')

plt.show()
```

```python
#red color cluster => [24.0, 45.714285714285715] => higher y value compared to x value
#green color cluster => [52.833333333333336, 61.833333333333336] => higher x value compared to y value
#blue color cluster =>  [62.4, 14.2] => close x and y value
```

**Scikit Learn for KMeans CLustering**

```python
import pandas as pd
import numpy as np

df=pd.DataFrame({'x':[12,24,28,33,18,29,52,45,24,55,51,61,53,69,72,64,49,58],
                  'y':[36,39,30,52,54,46,55,59,63,70,66,63,58,23,14,8,19,7]})

df
```

```python
from sklearn.cluster import KMeans
model1 = KMeans(n_clusters=3)

model1.fit(df)
```

```python
df['label'] = model1.predict(df)
```

```python
model1.cluster_centers_
```

```python
df.head()
```

```python
plt.scatter(df.x,df.y,c=df.label)
```
